# "Construction of a drought monitoring model using deep learning based on multi-source remote sensing data" by Shen et al.
## Method
- https://doi.org/10.1016/j.jag.2019.03.006

- Regression
  - Deep Neural network
    - 500 neurons
    - min. 1000 epochs training
- Multiple value as input, SPEI as output


## Used data
### Remote sensing data
- MOD11A2
- MOD13A3
- MCD12Q1
- DEM
- TRMM3B43
### Soil data
- Soil texture data
  
## Used remote sensing indices
- VCI (Vegetation Condition Index)
- TCI (Temperature Condition Index)
- VSWI (Vegetation Supply Water Index)
- DEM ([Shuttle Radar Topography Mission] Digital Elevation Model)
- TRMM-Z (Tropical Rainfall Measuring Mission)
- Pa (Precipitation Anomaly)
- LC (Land Coverage)
- AWC (Available Water Capacity)
